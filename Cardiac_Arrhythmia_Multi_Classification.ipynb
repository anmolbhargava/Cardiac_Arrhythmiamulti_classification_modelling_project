### Handling the missing values

```python
df.isnull().values.any()
```

```python
missing_df = df.isnull().sum()[df.isnull().sum() > 0].reset_index()
missing_df.columns = ['column_name', 'missing_count']
missing_df
```

```python
missing_df = missing_df.sort_values(by='missing_count')

ind = np.arange(missing_df.shape[0])
width = 0.9
fig, ax = plt.subplots(figsize=(8,6))
rects = ax.barh(ind, missing_df.missing_count.values, color='blue')
ax.set_yticks(ind)
ax.set_yticklabels(missing_df.column_name.values, rotation='horizontal')
ax.set_xlabel("Count of missing values")
ax.set_title("Number of missing values in each column")
plt.show()
```

#### Column 'J' is dropped as it has 83.2% values missing

```python
df = df.drop('J', axis=1)
```

#### Deleting rows with missing values in "QRST" & "Heart_Rate"

```python
df = df.dropna(subset=["QRST","Heart_Rate"])
df.shape
```

```python
df.isnull().sum()[df.isnull().sum() > 0].reset_index()
```

#### Checking for outliers

```python
fig = plt.figure(figsize=(12,8))
df[["T","P"]].boxplot();
```

#### Since there are many outliers in the dataset, median is used for imputation.

```python
from sklearn.preprocessing import Imputer
columns_to_impute = ['T', 'P']
df[columns_to_impute] = Imputer(missing_values = np.nan, strategy = 'median', axis=0).fit_transform(df[columns_to_impute])
```

#### One final check for missing values

```python
df.isnull().values.any()
```

#### Removing the classes with less than 5 observations

```python
df = df[(df['Class_Nom'] != 7) & (df['Class_Nom'] != 8) & (df['Class_Nom'] != 14)]
```

```python
df['Class_Nom'].value_counts()
```

```python
df.shape
```

#### Removing the constant columns (columns with only one value)

### Models

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

#### Data Splitting into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

#### Min-Max Scaling (Data Standardization/ Normalization)

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns = X_train.columns )
X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns = X_test.columns)
```

```python
X_train_scaled.describe()
```

## ## KNN

#### KNN - GridSearch using CV

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
k = np.arange(15) + 1
param_knn = [{'n_neighbors' : k}]
knn_clf = GridSearchCV(knn, param_knn, cv=5)
knn_clf.fit(X_train_scaled, y_train)
```

#### KNN - Best Parameters from Grid Search

```python
knn_clf.best_params_
```

#### KNN - Training and Testing data accuracies

```python
KNN_best = KNeighborsClassifier(n_neighbors=3).fit(X_train_scaled, y_train)

print('Accuracy of KNN classifier on training set: %.2f' % (KNN_best.score(X_train_scaled, y_train)))
print('Accuracy of KNN classifier on testing set: %.2f' % (KNN_best.score(X_test_scaled, y_test)))
```

#### KNN - Cross Validation accuracies

```python
scores = cross_val_score(knn, X_train_scaled, y_train, cv=5)
print("Cross validation scores: %s" % (scores))
print("Average cross-validation scroe: %.2f" % (scores.mean()))
```

#### KNN - Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

KNN_predicted = KNN_best.predict(X_test_scaled)

y_actu = pd.Series(np.array(y_test['Class_Nom']), name='Actual')
y_pred = pd.Series(KNN_predicted, name='Predicted')
df_confusion = pd.crosstab(y_actu, y_pred, margins=True)
df_confusion
```

#### KNN - Classification accuracies report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, KNN_predicted))
```

#### KNN - Bagging

```python
from sklearn.ensemble import BaggingClassifier

bag_knnclf = BaggingClassifier(KNN_best, n_estimators=100, max_samples=1.0, bootstrap=True, n_jobs=-1, random_state=10)

bag_knnclf.fit(X_train_scaled, y_train)

print('Train score: %.2f' % (bag_knnclf.score(X_train_scaled,y_train)))
print('Test score: %.2f' % (bag_knnclf.score(X_test_scaled,y_test)))
```

```python
print(classification_report(y_test, bag_knnclf.predict(X_test_scaled)))
```

#### Bagging reduced the testing accuracy compared to original KNN classifier, but it improved training accuracy.

#### KNN - Pasting

```python
past_knnclf = BaggingClassifier(KNN_best, n_estimators=100, bootstrap=False, n_jobs=-1, max_samples=1.0, random_state=10)

past_knnclf.fit(X_train_scaled, y_train)

print('Train score: %.2f' % (past_knnclf.score(X_train_scaled, y_train)))
print('Test score: %.2f' % (past_knnclf.score(X_test_scaled, y_test)))
```

```python
print(classification_report(y_test, past_knnclf.predict(X_test_scaled)))
```

#### Pasting neither reduced nor improved testing or training accuracies compared to original KNN classifier
